クラスタリングアルゴリズムの手順
ステップ1: 上を探索

現在の探索対象のピクセルの上にあるピクセルにフォーカスを当てます。
1-1: 上の隣接ピクセルが閾値を超えていれば、現在のピクセルを上のクラスターに統合し、次のサブステップ（1-2）に進みます。
1-2: 左の隣接ピクセルにフォーカスを変更し、左のピクセルが異なるクラスターに属していれば、左のクラスターを上のクラスターに統合します。その後、探索対象のピクセルを次のピクセルへ移動します。
ステップ2: 左を探索

現在の探索対象のピクセルの左にあるピクセルにフォーカスを当てます。
2-1: 左の隣接ピクセルが閾値を超えていれば、現在のピクセルを左のクラスターに統合し、次のサブステップ（2-2）に進みます。
2-2: 右上の隣接ピクセルにフォーカスを変更し、右上のピクセルが異なるクラスターに属していれば、右上のクラスターを左のクラスターに統合します。その後、探索対象のピクセルを次のピクセルへ移動します。
ステップ3: 左上を探索

現在の探索対象のピクセルの左上にあるピクセルにフォーカスを当てます。
3-1: 左上の隣接ピクセルが閾値を超えていれば、現在のピクセルを左上のクラスターに統合し、次のサブステップ（3-2）に進みます。
3-2: 右上の隣接ピクセルにフォーカスを変更し、右上のピクセルが異なるクラスターに属していれば、右上のクラスターを左上のクラスターに統合します。その後、探索対象のピクセルを次のピクセルへ移動します。
ステップ4: 右上を探索

現在の探索対象のピクセルの右上にあるピクセルにフォーカスを当てます。
4-1: 右上の隣接ピクセルが閾値を超えていれば、現在のピクセルを右上のクラスターに統合します。その後、探索対象のピクセルを次のピクセルへ移動します。
ステップ5: 新しいクラスターの作成


```python
import numpy as np

# CSVファイルからデータを読み込む（仮のファイルパスを使用）
data_avg = pd.read_csv('csvdata_avg.csv', header=None)

# 指定された範囲のデータを抽出
data = data_avg_transposed.iloc[y_start:y_end, x_start:x_end].values

# 閾値
threshold = avg_event_mean+3*avg_event_stddev



# クラスター情報を格納する辞書
clusters = {}
current_cluster_id = 0

# データグリッドをループ処理
for x in range(data.shape[0]):
    for y in range(data.shape[1]):
        # 現在のピクセルが閾値を超えている場合のみ処理
        if data[x, y] > threshold:
            cluster_found = False

            # ステップ1: 上を探索
            if x > 0 and data[x-1, y] > threshold:
                # 1-1: 上の隣接ピクセルが閾値を超えていれば統合
                clusters[(x, y)] = clusters.get((x-1, y), current_cluster_id)
                cluster_found = True

                # 1-2: 左のピクセルを確認
                if y > 0 and data[x, y-1] > threshold:
                    left_cluster_id = clusters.get((x, y-1))
                    if left_cluster_id is not None and left_cluster_id != clusters[(x, y)]:
                        # 左のクラスターを上のクラスターに統合
                        for key, value in clusters.items():
                            if value == left_cluster_id:
                                clusters[key] = clusters[(x, y)]

            # ステップ2: 左を探索
            if y > 0 and data[x, y-1] > threshold and not cluster_found:
                # 2-1: 左の隣接ピクセルが閾値を超えていれば統合
                clusters[(x, y)] = clusters.get((x, y-1), current_cluster_id)
                cluster_found = True

                # 2-2: 右上のピクセルを確認
                if x > 0 and data[x-1, y+1] > threshold:
                    top_right_cluster_id = clusters.get((x-1, y+1))
                    if top_right_cluster_id is not None and top_right_cluster_id != clusters[(x, y)]:
                        # 右上のクラスターを左のクラスターに統合
                        for key, value in clusters.items():
                            if value == top_right_cluster_id:
                                clusters[key] = clusters[(x, y)]

            # ステップ3: 左上を探索
            if x > 0 and y > 0 and data[x-1, y-1] > threshold and not cluster_found:
                # 3-1: 左上の隣接ピクセルが閾値を超えていれば統合
                clusters[(x, y)] = clusters.get((x-1, y-1), current_cluster_id)
                cluster_found = True

                # 3-2: 右上のピクセルを確認
                if data[x-1, y+1] > threshold:
                    top_right_cluster_id = clusters.get((x-1, y+1))
                    if top_right_cluster_id is not None and top_right_cluster_id != clusters[(x, y)]:
                        # 右上のクラスターを左上のクラスターに統合
                        for key, value in clusters.items():
                            if value == top_right_cluster_id:
                                clusters[key] = clusters[(x-1, y-1)]

            # ステップ4: 右上を探索
            if x > 0 and y < data.shape[1] - 1 and data[x-1, y+1] > threshold and not cluster_found:
                # 4-1: 右上の隣接ピクセルが閾値を超えていれば統合
                clusters[(x, y)] = clusters.get((x-1, y+1), current_cluster_id)
                cluster_found = True

            # ステップ5: 新しいクラスターの作成
            if not cluster_found:
                clusters[(x, y)] = current_cluster_id
                current_cluster_id += 1


# クラスターのサイズを計算
cluster_sizes = {}
for cluster_id in clusters.values():
    cluster_sizes[cluster_id] = cluster_sizes.get(cluster_id, 0) + 1

# サイズが5以上のクラスターのみを保持
filtered_clusters = {pos: cid for pos, cid in clusters.items() if cluster_sizes[cid] >= 5}

    
for (x, y), cluster_id in filtered_clusters.items():
    absolute_x = y + x_start
    absolute_y = x + y_start
    print(f"ピクセル ({absolute_x}, {absolute_y}) はクラスター {cluster_id} に属します。")    
    
   

    
    
```

```python

    
# フィルタリングされたクラスターの数
num_filtered_clusters = len(set(filtered_clusters.values()))

# 各フィルタリングされたクラスターに含まれるピクセル数とデータの合計値
filtered_cluster_pixel_counts = {}
filtered_cluster_sum_values = {}
filtered_cluster_max_values = {}
filtered_cluster_max_positions = {}

for (x, y), cluster_id in filtered_clusters.items():
    filtered_cluster_pixel_counts[cluster_id] = filtered_cluster_pixel_counts.get(cluster_id, 0) + 1
    filtered_cluster_sum_values[cluster_id] = filtered_cluster_sum_values.get(cluster_id, 0) + data[x, y]
    
    # 各クラスターの最大値とその座標を更新
    current_value = data[x, y]
    if cluster_id not in filtered_cluster_max_values or current_value > filtered_cluster_max_values[cluster_id]:
        filtered_cluster_max_values[cluster_id] = current_value
        filtered_cluster_max_positions[cluster_id] = (x, y)

        
        
        
# 結果の表示
print(f"フィルタリングされたクラスターの数: {num_filtered_clusters}")
print("各フィルタリングされたクラスターに含まれるピクセル数とデータの合計値:")
for cluster_id in filtered_cluster_pixel_counts:
    print(f"クラスター {cluster_id}: ピクセル数 = {filtered_cluster_pixel_counts[cluster_id]}, 合計値 = {filtered_cluster_sum_values[cluster_id]}")

filtered_cluster_variance_sum = {}
filtered_cluster_stddev = {}

for (x, y), cluster_id in filtered_clusters.items():
    # 各ピクセルの標準偏差の二乗を合計
    stddev_squared = extracted_array_std[x, y] ** 2
    filtered_cluster_variance_sum[cluster_id] = filtered_cluster_variance_sum.get(cluster_id, 0) + stddev_squared

# 平均分散（平均二乗変動）を計算し、その平方根を取る
for cluster_id in filtered_cluster_variance_sum:
    # 平均分散 = 合計分散 / ピクセル数
    average_variance = filtered_cluster_variance_sum[cluster_id] / filtered_cluster_pixel_counts[cluster_id]
   
    # 標準偏差 = 平均分散の平方根
    filtered_cluster_stddev[cluster_id] = np.sqrt(average_variance)

print(filtered_cluster_max_positions[4])   
```

```python
# 各クラスターに含まれる座標のリストを作成
filtered_cluster_coordinates = {}

for (x, y), cluster_id in filtered_clusters.items():
    if cluster_id not in filtered_cluster_coordinates:
        filtered_cluster_coordinates[cluster_id] = []
    filtered_cluster_coordinates[cluster_id].append((y+x_start, x+y_start))

# 各クラスターに含まれる座標を表示
for cluster_id, coordinates in filtered_cluster_coordinates.items():
    print(f"クラスター {cluster_id}: 座標 = {coordinates}")
    
# すべてのクラスターの座標を一つのリストに統合
list_coordinates = []
for coordinates in filtered_cluster_coordinates.values():
    list_coordinates.extend(coordinates)

# 結果の表示
print("すべてのクラスターの座標:")
print(list_coordinates)
```
